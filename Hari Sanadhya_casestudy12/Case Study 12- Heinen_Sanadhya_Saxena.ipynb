#### 3.1 Activation function in Neural Network and its examples
In a neural network, the artificial neurons calculate a weighted sum of its input, adds a bias and then decides whether it should be fired or not. This firing of neuron is decided by using the mathematical functions named as Activation Functions. Since the computation of weighted sum and addition of bias is a linear process (y = wx + b), the activation functions used are non-linear functions so that the neurons can learn complex relationship between the input and the output.
Below is the simple summary of some of the activation functions (keras activations documentation at https://keras.io/activations/ used as a reference here) showing their representation:

-   Sigmoid - Output is in the range of 0 to 1 centered at 0.5 (output is 0.5 when x=0). The mathematical form of this function is 
<center>$f(x){=}{\frac{1}{1+e^{-x}}}$</center>


- tanh - The output of this function is in the range of -1 to 1 centered at 0 (output is 0 when x=0). 
<center>$f(x){=}{\frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}}$</center>


-	Rectified Linear Unit (RELU) – It returns element-wise max(x, 0) i.e. x when x>0 else 0.


-	Exponential Linear Unit (ELU) – Returns <br>&nbsp;&nbsp;&nbsp;x when x > 0 <br>&nbsp;&nbsp;&nbsp;else returns ${alpha}*{(e^x-1)}$ <br>where alpha is a predefined constant.


-	Scaled Exponential Linear Unit (SELU) – Returns scale * elu(x, alpha), where alpha and scale are pre-defined constants. The values of alpha and scale are chosen so that the mean and variance of the inputs are preserved between two consecutive layers.


-	Softplus activation function (softplus) – Returns  $log(e^x + 1)$.


-	LeakyReLU (Leaky version of a Rectified Linear Unit) – Returns <br>&nbsp;&nbsp;&nbsp;alpha * x when x < 0 <br>&nbsp;&nbsp;&nbsp;else x <br>where aplha is a very small predefined value allowing a small gradient when the regular relu unit is not active (x<0).


-	Rectified Linear 6 (RELU6) – It imposes an upper limit of 6 on the relu output. Returns <br>&nbsp;&nbsp;&nbsp;6 when x>=6 <br>&nbsp;&nbsp;&nbsp;else return x when x>=0 <br>&nbsp;&nbsp;&nbsp;else returns 0. <br>


#### 3.2 Keras kernal and bias initializers
Initializations define a way to set the initial random weights of Keras layers. For a neural network layer, both the weights and the bias of the neuron can be initialized. The better this initialization is, the less time it takes for the neural network to reach the global minimum loss. Bias is mostly initialized to zero. Only the weights are initialized to the random values (parameter - kernel_initializer). Below are some of the methods to find these random values (reference - https://keras.io/initializers/):

-	LeCun uniform initializer (lecun_uniform) – It draws samples from a uniform distribution within [-limit, limit] where limit is sqrt(3 / fan_in) where  fan_in is the number of input units in the weight tensor.
-	Glorot uniform initializer, also called Xavier uniform initializer (glorot_uniform) – It draws samples from a uniform distribution within [-limit, limit] where limit is sqrt(6 / (fan_in + fan_out)) where fan_in is the number of input units in the weight tensor and fan_out is the number of output units in the weight tensor.
-	He uniform variance scaling initializer (he_uniform) – It draws samples from a uniform distribution within [-limit, limit] where limit is sqrt(6 / fan_in) where  fan_in is the number of input units in the weight tensor.<br>
-   Glorot normal initializer (glorot_normal) - It draws samples from a truncated normal distribution centered on 0 with  stddev = sqrt(2 / (fan_in + fan_out)) where fan_in is the number of input units in the weight tensor and fan_out is the number of output units in the weight tensor.


#### 3.3 Keras Model Optimizer
Loss function in neural network is a mathematical way of measuring how wrong the model predictions are. The model optimizer functions to tie together the loss function and model parameters by updating the model in response to the output of the loss function so as to try and minimize the loss function and thereby make model predictions as correct as possible. Optimizers shape and mold the neural network model into its most accurate possible form by futzing with the weights. The loss function is the guide to the terrain, telling the optimizer when it’s moving in the right or wrong direction. Some of the optimizers that we have used in this case study includes (reference - http://ruder.io/optimizing-gradient-descent/index.html#stochasticgradientdescent and https://keras.io/optimizers):

- SGD - Stands for Stochastic Gradient Descent optimizer. In order to move towards the minimum loss, instead of using all the observation in the entire batch to compute the gradient (gradient descent optimizer), this function uses a single randomly chosen example from the current batch to compute the gradient. So for SGD to work, the number of iterations required is large but its computationally very cheap and fast though noisy.

- RMSprop - Its a variation of the mini-batch gradient descent where the learning rate for a weight is divided by its magnitude. So its mini-batch version to use just the sign of the gradient. 

- Adadelta - This optimizer adapts learning rates based on a moving window of gradient updates, instead of accumulating all past gradients. This is the reason that for this optimizer, its not required to set an initial learning rate.

- Adam - Stands for Adaptive Moment Estimation. It is an update to the RMSProp optimizer. In this optimization algorithm, running averages of both the gradients and the second moments of the gradients are used. 

- Adamax - AdaMax is an adaptive stochastic gradient descent method and a variant of Adam based on the infinity norm. 


#### 3.4 Some useful Neural terminology
- **one epoch** refers to one forward pass and one backward pass of all the training examples
- **batch size** refers to the number of training examples in one forward/backward pass. The higher the batch size, the more memory space will be consumed and the less the batch size, the slower will be the learning process.
- **number of iterations** refers to number of forward/backward passes that a neural network takes to complete one epoch. Mathematically it can be said as 
<center> $\#\_of\_iterations{=}{ceil({\frac{number\_of\_training\_examples}{batch\_size}})}$  where ceil is the math ceiling function</center>

### Answer to Theory Question 1
**Question 1:** What was the effect of adding more layers/neurons.<br>
**Solution:** A very basic neural network which is a network with single hidden layer (Model 7) has just 29 paramters to be trained. On training, the model gaves an ROC score of 0.68 with accuracy on test data as 64%. On increasing number of layers from one to 2, the model accuracy got better. Model 6 had 28 neurons in first layer and one neuron in second layer gave ROC score of 0.79. Model 5 which is similar to model 6 with the only difference as 140 neurons in the first layer resluted in ROC score of 0.80 which is a little better then the model 6. Comparing the results of model 5 and 6 with the models 1, 2, 3 and 4 which are having more trainable parameters and larger number of hidden layers, the ROC score and accuracy decreased. As observed from table 1 above, the accuracy get decreasing as we make the model deeper. This happens because of the activation function - sigmoid. Since the sigmoid function returns value between 0 and 1 and the dataset values are already scaled, the output from sigmoid are small values (small values of weight and bias). So as we go deep, the overall neural network learning decreases since deeper layers were not able to learn much of the relationship between the input and output features. So using just the sigmoid function, its better to use a neural network with just 2 hidden layers with high number of neurons in the first layer to get better performing models.

Second hidden layer has a dropout of 10% and the last hidden layer is dense layer with single neuron followed by sigmoid activation function. ReLU has advantages that includes:
* non-saturation of its gradient that results in avoiding vanishing gradients problem, which greatly accelerates the convergence of stochastic gradient descent  
* less computationally expensive as it uses simple math functions as compared to sigmoid and tanh
* it runs much faster than sigmoid function
Apart from ReLU, using dropout prevents the network from overfitting. SGD optimizer in turn adds to the fast learning. Though adding number of dense layers improved the model performance, but since that increase was not that significant, the model chosen was the one with good enough performance. 
Uniform kernel initializer gave best results and the reason for that in our understanding is bucause of the dataset being standardized, uniform initializer initializes the weights uniformally between the minimum and maximum values of 0 and 1 (default values of minval and maxval parameters for uniform kernel initializer)

```python
# Compute ROC curve and ROC area
from sklearn.metrics import roc_curve, auc
def plot_roc_binary_classification(y_actual, y_predicted):
    fpr, tpr, _ = roc_curve(y_actual, y_predicted)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
    plt.xlim([0.0, 1.00])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# Compute ROC curve and ROC area
from sklearn.metrics import roc_curve, auc
def plot_roc_binary_classification(y_actual, y_predicted):
    fpr, tpr, _ = roc_curve(y_actual, y_predicted)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
    plt.xlim([0.0, 1.00])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
```
