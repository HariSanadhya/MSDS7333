## 1. Abstract

<p style="text-align:justify">Data comes in many forms, most of it messy. Messy refers to state of the data which cannot be used in the same form as it exists without impacting the quality of the data analysis/modelling. Messy data includes missing data, unstructured data, or data that lacks regular structure. In case of messy data, cleansing of the data is a prerequisite before working with it to improve the quality in the value that is extracted from the data.<br><br>Real world datasets are often quite messy and not well-organized for available data analysis tools. Data Scientists land up spending most of their time with preprocessing of the data which includes cleansing of the dataset and feature generation. One of the major problem with messy data is the problem of missing data, data can be unavailable for any number of reasons. Sometimes study participants fail to complete all the items on a survey, other times there are problems with data recording.The impact of missing data can reduce the statistical power of a study and can produce biased estimates, leading to invalid conclusions. How researchers or analysts choose to handle those data can be tricky business. <br><br>In this case study, we focus on the types of missingness in the data and the ways to handle the missingness so that the quality is not impacted much due to missing data.</p>

## 2. Introduction

<p style="text-align:justify">The concept of missing values is important to understand in order to successfully manage data.  If the missing values are not handled properly, then inaccurate inference about the data may be drawn. In statistics, missing data, or missing values, occur when no data value is stored for the variable in an observation. Missing data are a common occurrence and can have a significant effect on the conclusions that can be drawn from the data. Missing data present various problems like</p>

* the absence of data reduces statistical power of a statistical test 
* it can cause bias in the estimation of parameters. 
* it can reduce the representativeness of the samples. 
* and many more...

<p style="text-align:justify">Each of these problems may threaten the validity of the analysis and can lead to invalid conclusions. It becomes important to handle the missing data issues and address it before we start our reserch or analysis. There are different kinds of missingness and ways in which the missing data is handled.<br><br>The purpose of this study is to examine and analyze the different types of missingness existing in the dataset and then handle the missingness before training the machine learning model. The results are then compared with one another as conclusion of the case study.<\p>

## 3. Backgroud



Missing data (or missing values) is defined as the data value that is not stored for a variable in the observation of interest. Before we review the methods of data imputation, we have to understand the reason why data goes missing.

#### Types of Missing Data
In general, there are three types of missing data according to the mechanisms of missingness.
* <p style="text-align:justify">Missing completely at random (MCAR): MCAR is defined as when the probability that the data are missing is not related to either the specific value which is supposed to be obtained or the set of observed responses. MCAR refers to the propensity for a data point to be missing is completely random i.e. there is no correlation between missing and non-missing data points. For example - if data are missing by design, because of an equipment failure or because the samples are lost in transit or technically unsatisfactory, such data are regarded as being MCAR. The statistical advantage of data that are MCAR is that the analysis remains unbiased. Power may be lost in the design, but the estimated parameters are not biased by the absence of the data.</p>

* <p style="text-align:justify">Missing at random (MAR): Data are regarded to be MAR when the probability that the responses are missing depends on the set of observed responses, but is not related to the specific missing values which is expected to be obtained. For example, consider an dataset where data is missing in 2 different columns (X and Y). On analysis it appears that data is missing whenever a third variable (Z) is having some specific value i.e some condition like when Variable Z > 30, then X and Y have missing data but that missingness is random. </p>

* <p style="text-align:justify">Missing not at random (MNAR): If the characters of the data do not meet those of MCAR or MAR, then they fall into the category of missing not at random (MNAR). This is case when missingness is having some pattern. The cases of MNAR data are problematic. The only way to obtain an unbiased estimate of the parameters in such a case is to model the missing data. The model may then be incorporated into a more complex one for estimating the missing values. </p>

### Ways to handle missing data
Many methods have been proposed to handle missing data. Some of the most popular onces include:
* <p style="text-align:justify">Deletion: One of the most common way to handle the missing data is to delete or omit the records having missing data from the study. The deletion can be listwise or pairwise deletion. </p>

    * <p style="text-align:justify">Listwise deletion refers to simply omit the cases/records with the missing data and analyze the remaining data and for that reason is also known as the complete case (or available case) analysis. This is the default option of most of the statistical software packages. When sample size is very small, or missing data is not MCAR, the listwise deletion is not the optimal strategy as it may cause bias in parameter estimates. </p>

    * <p style="text-align:justify">Pairwise deletion eliminates information only when the particular data-point needed to test a particular assumption is missing. Pairwise deletion attempts to minimize the loss that occurs in listwise deletion.  An easy way to think of how pairwise deletion works is to think of a correlation matrix.  A correlation measures the strength of the relationship between two variables.  For each pair of variables for which data is available, the correlation coefficient will take that data into account.  Thus, pairwise deletion maximizes all data available by an analysis by analysis basis.</p>


* <p style="text-align:justify">Imputation: Imputation is replacing missing values with substitute values. The following methods use some form of imputation.</p>

    * <p style="text-align:justify">Educated Guessing: It sounds arbitrary and isn’t your preferred course of action, but can be used as a way to handle missing data by inferring a missing value via educated guess i.e. by using experience or by looking at data or by contacting SME's to make the guess of the value that can be used in place of the missing value.</p> 

    * <p style="text-align:justify">Average Imputation: Use the average value of the responses from the other participants to fill in the missing value. If the average of the 30 responses on the question is a 4.1, use a 4.1 as the imputed value. This choice is not always recommended because it can artificially reduce the variability of your data but in some cases makes sense.   </p>
    * <p style="text-align:justify">Common-Point Imputation: For a rating scale, using the middle point or most commonly chosen value which means substitution using mode or median values. For example, on a five-point scale, substitute a 3, the midpoint, or a 4, the most common value (in many cases). This is a bit more structured than guessing, but it’s still among the more risky options. </p>
    * <p style="text-align:justify">Regression Substitution: You can use multiple-regression analysis to estimate a missing value. Regression substitution predicts the missing value from the other values which are not missing for the same record.</p>
    * <p style="text-align:justify">Multiple Imputation: The most sophisticated and, currently, most popular approach is to take the regression idea further and take advantage of correlations between responses. In multiple imputation (MICE and MCMC), software creates plausible values based on the correlations for the missing data and then averages the simulated datasets by incorporating random errors in your predictions. It is one of a number of examples where computers continue to change the statistical landscape. </p>


* Other methods:
    * <p style="text-align:justify">Dummy Variable approach: This method if the missing data in continuous creates of a new feature which will indicate whether the data was missing or not for a particular feature (eg 1 when missing and 0 when not missing) and then the inputation is done on the missing value. If the missing data is categorical, then a new category indicating missing value is created and all the missing values for that column are replaced with that value.</p>
    * <p style="text-align:justify">Some state-of-the-art imputation techniques including Singular Value Decomposition (SVD) and K-Nearest Neighbors (KNN) based methods which usually achieve good performances, but are computationally expensive especially for large datasets. </p>
    * <p style="text-align:justify">Multiple Imputation using Denoising Autoencoders (MIDA) and  Denoising Autoencoder with Partial Loss (DAPL) are two deep learning based alternative for missing data imputation. </p>

#### 4.2 Training and Evaluation of Baseline Model
Baseline model is the trained linear regression model on the data which is the outcome of the Exploratory Data Analysis performed in step 4.1 above. No Cross-Validation is performed as generalization is not the target of the case study but the target is to compare the change in the results of the models as the data has different types of missing data in it. Evaluation matrics include loss and goodness of fit parameters.

#### 4.3 Missing data insertion and Imputation
The dataset used to create the baseline model does not have any missing values. So in this case study, we created new datasets from the actual dataset and inserted missing values into these datasets. 10 copies of the dataset were created, the description of those are as below:-

1. 1 percent missing values in the column NOX such that missingness is Missing Completely At Random (MCAR)

2. 5 percent missing values in the column NOX such that missingness is Missing Completely At Random (MCAR)

3. 10 percent missing values in the column NOX such that missingness is Missing Completely At Random (MCAR)

4. 20 percent missing values in the column NOX such that missingness is Missing Completely At Random (MCAR)

5. 33 percent missing values in the column NOX such that missingness is Missing Completely At Random (MCAR)

6. 50 percent missing values in the column NOX such that missingness is Missing Completely At Random (MCAR)

7. Total 10 percent missing values in the columns 'INDUS' and 'PTRATIO' with randomness contolled partially by NOX column values. This case is Missing At Random (MAR). Either INDUS or PTRATIO or both have missing values whenever NOX value is less than NOX's 10 percentile value.

8. Total 20 percent missing values in the columns 'INDUS' and 'PTRATIO' with randomness contolled partially by NOX column values. This case is Missing At Random (MAR). Either INDUS or PTRATIO or both have missing values whenever NOX value is less than NOX's 20 percentile value.

9. Total 30 percent missing values in the columns 'INDUS' and 'PTRATIO' with randomness contolled partially by NOX column values. This case is Missing At Random (MAR). Either INDUS or PTRATIO or both have missing values whenever NOX value is less than NOX's 30 percentile value.

10. 25 percent missing values in the column 'INDUS' with randomness having a dependent pattern with NOX column values. This case is Missing Not At Random (MNAR). INDUS is having missing values whenever NOX value is less then NOX's 25 percentile value.

In all the 10 cases above, we performed the imputation using the mean value of the column with missing data.


#### 4.4 Training and Evaluation of Liear Regression Model created on the Imputed datasets and comparison with baseline
All the 10 datasets created in the step 4.3 were then used to train linear regression models. Evaluation matrics include loss and goodness of fit parameters for all the different models trained are computed and then the comparison of these model metrics with performed with the baseline model.

## 6. Conclusion and Future work


As seen in Figure 27, the test results summary indicates very minor differences in the performance of the Linear Regression model when trained with data with and without imputation (baseline model). Firstly this dataset is small and for MAR and MNAR datasets, the columns containing missing values were not that impartant in making the prediction so even though a high percentage of those values were made missing and then imputed with mean, model does not show much variance in the results. For MCAR, the variable that was chosen to insert missing value, though had high importance (large negative coefficient) in prediction, the result still showed model with almost similar performance. This might be because the dataset is small and secondly, the NOX variable is almost normally distributed (Figure 5) and the random indexes chosen is also using gaussian distribution. This combination might have made the resultant distribution of the NOX column with missing value to follow the same distribution. Replacing the missing values with mean causes the bell curve of normal distribution to get more densed in the middle (mean lies at the center for normal distribution). So in case when the missing data was small, the linear regression coefficient for NOX column was not impacted much but as the missing percentage was increased, the value of the coefficient changed (from -13 in baseline model to 3.86 in the dataset with 50% values of NOX column missing). This means that though the replacement with mean may be a good option for missing data imputation for small percentage of missing values, it is not that good if the percentage of missing values is large and the variable with missing values is an important feature in decision making (prediction or classification). This is the reason that for missing data imputation, there are many methods which can be applied and this area is still a hot area of research. 

Future work would be to expand the analysis to a larger data set and test the performance and generalization capability of the model trained using missing data imputed with different imputation methods.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, r2_score as r2

def plot_actual_vrs_predicted(y_test_actual, y_test_predicted):
    """
    Function to create scatter plot of actual vrs predicted values of the target variable. Ideally all the points 
    should be on the line y=x (diagonal line)
        
    Arguments:
    y_test_actual -- Series containing the actual values of the target variable for test data
    y_test_predicted -- Series containing the predicteed values of the target variable for test data
       
    Returns:
    Does not return anything.
    """
    fig = plt.figure(figsize=(6, 6))
    plt.grid(color='g', linestyle='dotted', linewidth=1)
    plt.scatter(y_test_actual, y_test_predicted, s=2)
    plt.title("Actual vs Predicted prices")
    plt.xlabel("Actual Prices: $Y_i$")
    plt.ylabel("Predicted prices: $\hat{Y}_i$")
    max_value_of_y = int(max(y_test_actual.tolist() + y_test_predicted.tolist()))
    max_val_of_axis = max_value_of_y + 5
    plt.xlim((0,max_val_of_axis))
    plt.ylim((0,max_val_of_axis))
    plt.gca().set_aspect('equal')
    plt.xticks(np.arange(0, ((max_val_of_axis)/10)*10, 10))
    plt.yticks(np.arange(0, ((max_val_of_axis)/10)*10, 10))
    plt.plot([0,max_val_of_axis], [0,max_val_of_axis], ls="dotted", c="r", linewidth=1)
    plt.show()
    
def plot_residual(y_test_actual, y_test_predicted):
    """
    Function to create residual plot which is "for each test observation, plot the difference between actual and 
    predicted values". Ideally it should be the line y=0.
        
    Arguments:
    y_test_actual -- Series containing the actual values of the target variable for test data
    y_test_predicted -- Series containing the predicteed values of the target variable for test data
       
    Returns:
    Does not return anything.
    """
    plt.figure(figsize=(7, 6))
    plt.grid(color='g', linestyle='dotted', linewidth=1)
    plt.plot(y_test_actual - y_test_predicted, markersize=1, linewidth=1)
    plt.xlabel("Test Observation")
    plt.ylabel("Residual value \n(Actual - Predicted)")
    plt.title("Residual Plot")
    plt.show()

def fit_evaluate_linear_regression(train_x, test_x, train_y, test_y, feature_names) -> LinearRegression:
    """
    Given the train and test data, this method fits the train data to linear regression and then plots the results 
    and computes the regression matrics that can be used to compare various  
        
    Arguments:
    train_x -- Dataframe that contains the dependent variables for the train data
    test_x -- Dataframe that contains the dependent variables for the test data
    train_y -- Series containing the response variable for the train data
    test_y -- Series containing the response variable for the test data
    feature_names -- List containing the column headings of all the dependent variables that too in the column order as in the train_x OR test_x dataframes.  
       
    Returns:
    lm -- object of the LinearRegression class i.e. this function returns the trained LinearRegression model
    """
    
    lm = LinearRegression()
    lm.fit(train_x, train_y)
    pred_y = lm.predict(test_x)
    print()
    plot_actual_vrs_predicted(test_y, pred_y)
    print()
    plot_residual(test_y, pred_y)
    print()
    # print the coefficients
    print("Intercept of Linear regression:" + str(lm.intercept_))
    coefficient = []
    [coefficient.append("%.4f" % i) for i in lm.coef_]

    print()
    print("Coeffficient of every linear regression feature")

    print(pd.DataFrame(coefficient, columns=['Coefficient'], index=feature_names))

    printmd("**Linear regression metrics**")
    print("Mean Absolute Error: " + str(mae(test_y, pred_y)))
    print("Mean Squared Error: " + str(mse(test_y, pred_y)))
    print("Root Mean Squared Error: " + str(np.sqrt(mse(test_y, pred_y))))
    printmd("Goodness of fit ($R^2$): " + str(r2(test_y, pred_y)))
    
    return lm
```

```python
from sklearn.model_selection import train_test_split
def complete_linear_regression(x_data, y_data, feature_names, test_size=0.33, random_state = 5) -> LinearRegression:
    """
    This method splits the train and test data using the seed and test data size specified in the parameters and then 
    fits linear regression on the train data and evaluated model performance on test data
        
    Arguments:
    x_data -- Dataframe that contains the dependent variables
    y_data -- Series containing the response variable corresponding to the dependent variables
    feature_names -- List containing the column headings of all the dependent variables that too in the column order as 
        in the train_x OR test_x dataframes.  
    test_size -- Can be float, int, optional. If float, should be between 0.0 and 1.0 and represent the proportion
        of the dataset to include in the test split. If int, represents the absolute number of test samples. By default, 
        the value is set to 0.33.
    random_state -- Possible values are int, RandomState instance or None, optional (default=5). If int, random_state is 
    the seed used by the random number generator; If RandomState instance, random_state is the random number generator;
    If None, the random number generator is the RandomState instance used by `np.random`.
    
    Returns:
    lm -- object of the LinearRegression class i.e. this function returns the trained LinearRegression model
    """
    x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = test_size, random_state = random_state)
    x_train.reset_index(drop=True, inplace=True)
    y_train.reset_index(drop=True, inplace=True)
    x_test.reset_index(drop=True, inplace=True)
    y_test.reset_index(drop=True, inplace=True)
    print("Shape of Train data features: " + str(x_train.shape))
    print("Shape of Test data features: " + str(x_test.shape))
    print("Shape of Train data Response variable: " + str(y_train.shape))
    print("Shape of Test data Response variable: " + str(y_test.shape))
    return fit_evaluate_linear_regression(x_train, x_test, y_train, y_test, feature_names)
```

```python
import random
# Insert missing data Completely at Random
def insertMissing_CR(df_data, columns, percent) -> pd.DataFrame:
    """
    This method inserts missing values completely at random. 
        
    Arguments:
    df_data -- Dataframe that contains the data in which missing values are to be inserted
    columns -- List of columns in the dataframe df_data in which the missing values are to be inserted 
    percent -- a float value between 0 and 1, both inclusive, specifying the percent of data that is to be 
                replaced by missing values
    
    Returns:
    df_data -- Dataframe containing missing values 
    """
    print("Insertion of " + str(int(100*percent)) + "% missing data into columns " + str(columns))
    
    for column in columns:
        print("Count before inserting missing Data into column " + column + ": " + str(df_data[column].count()))
        prop = int(df_data.shape[0] * percent) # prop contains the number of records equivalent to the percent specified
        random_ind = set() # Initialize an empty set
        
        # since the random number generated are generated by replacement, so the random index might be duplicate.
        # So used set to collect distinct values and 
        while len(random_ind) < prop: 
            # get random indexes of columns where missing data will be inserted
            random_ind.add(random.choice(range(df_data[column].shape[0]))) 
        df_data[column][random_ind] = np.NaN
        print("Count after inserting missing Data into column " + column + ": " + str(df_data[column].count()))
    return df_data
```

```python
# Insert missing data Missing at Random
def insertMissing_MR(df_data, missing_data_cols, condition) -> pd.DataFrame:
    """
    This method inserts missing values at random. 
        
    Arguments:
    df_data -- Dataframe that contains the data in which missing values are to be inserted
    missing_data_cols -- List of 2 columns in the dataframe df_data in which the missing values are to be 
                inserted. Two columns is necessary else the logic implemented needs to be modified 
    Condition -- A bool series which is of the length equal to the number of rows in the dataframe df_data. 
            Depending on the bool value, the rows in which missing values are to be inserted will be determined 
    
    Returns:
    df_data -- Dataframe containing missing values 
    """
    missing_col_index = df_data.index[condition].tolist()
    missing_cond = [0, 1, 2] 
    for col_index in missing_col_index:
        option = missing_cond[np.random.randint(0, np.size(missing_cond))]
        # If option = 0, then insert missing value into first column
        # If option = 1, then insert missing value into second column
        # If option = 2, then insert missing value into both the columns
        if option == 0:
            df_data[missing_data_cols[option]][col_index] = np.NaN
        elif option == 1:
            df_data[missing_data_cols[option]][col_index] = np.NaN
        else:
            df_data[missing_data_cols[0]][col_index] = np.NaN
            df_data[missing_data_cols[1]][col_index] = np.NaN
        
    for column in missing_data_cols:
        print("Count after inserting missing Data into column " + column + ": " + str(df_data[column].count()))
    missing_data_rows = sum(df_data.apply(lambda x: sum(x.isnull().values), axis = 1)>0)
    print("Number of rows with at least one missing value: " + str(missing_data_rows) + " which is " + 
          str(round(missing_data_rows/df_data.shape[0]* 100,2)) + "% of the entire data")
    return df_data
```

```python
# Insert missing data Missing not at Random
def insertMissing_NR(df_data, columns, condition):
    """
    This method inserts missing values not at random. There will be a pattern in missingness and that pattern is
    handled by the condition parameter.
        
    Arguments:
    df_data -- Dataframe that contains the data in which missing values are to be inserted
    columns -- List of columns in the dataframe df_data in which the missing values are to be inserted.
    Condition -- A bool series which is of the length equal to the number of rows in the dataframe df_data. 
            Depending on the bool value, the rows in which missing values are to be inserted will be determined 
    
    Returns:
    df_data -- Dataframe containing missing values 
    """
    missing_col_index = df_data.index[condition].tolist()
    
    for column in columns:
        before_count = df_data[column].count()
        print("Count before inserting missing Data into column " + column + ": " + str(before_count))
        df_data[column][missing_col_index] = np.NaN
        after_count = df_data[column].count()
        print("Count after inserting missing Data into column " + column + ": " + str(after_count))
        print("Missing data is: " + str(round((before_count - after_count)/before_count*100, 2)) + "%")
    return df_data
```
